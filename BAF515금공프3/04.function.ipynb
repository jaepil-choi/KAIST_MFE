```python
import numpy as np
```

```python
def count_odd(l):
    odd_count = 0
    for i in l:
        if i % 2 != 0:
            odd_count += 1
    
    return odd_count
```

```python
a = count_odd(l=[1,2,5,7,8])
a
```

Self-Exercise: Numerical Square Root

```python
def num_sqrt(n): #  . 
    # Initial Values
    epsilon = 0.00001
    
    r = n/2
    lin1 = n/r
    lin2 = r 

    while abs(lin1 - lin2) >= epsilon:
        r = (lin2 - lin1) / 2
        lin1 = n/r
        lin2 = r
    
    return r
```

```python
num_sqrt(9)
```

```python
num_sqrt(10000)
```
